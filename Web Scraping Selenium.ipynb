## Import the necessary library for web scraping:

```python
from selenium import webdriver
import time
import pandas as pd
```

### Setup Selenium Web-Driver:
in this step we have configuared chrome selenium web driver. so, required chromedriver.exe fie. Add this exe file as executable path and ready for further web scraping techniques.

### Download link for chrome driver exe file :
https://chromedriver.storage.googleapis.com/index.html?path=104.0.5112.29/

### website link address :
https://www.indeed.com/q-data-scientist-jobs.html

```python
from selenium.webdriver.chrome.options import Options
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")
search_query_page = 'https://www.indeed.com/q-data-scientist-jobs.html'
driver_path='F:/cli/Data_science_projects/webScrapping/selenium/chromedriver.exe'
driver_chrome_bar = webdriver.Chrome(executable_path=driver_path,options=options)
driver_chrome_bar.get(search_query_page)
driver_chrome_bar.maximize_window()
print("===print the page sources===",driver_chrome_bar.page_source)
driver_chrome_bar.maximize_window()
time.sleep(5)
```

#### Before going to the web scraping detailing, wejust check its syntax and how to target element through X_path 

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
job_list = driver_chrome_bar.find_element(By.XPATH,"//div[@class='job_seen_beacon']")
job_list

element = job_list.find_element(by=By.XPATH, value="//h2//a//span")
print(type(job_list))
print(element.get_attribute('innerHTML'))
print(job_list)
```

### Scrape data from the web-page from various web element and stored into an array :

```python
job_details = []
job_info = []

#list of all jobs with their elements
job_list = driver_chrome_bar.find_elements(By.XPATH,"//div[@class='job_seen_beacon']")
#job_list = driver_chrome_bar.find_elements(By.XPATH,"//div[@class='job_seen_beacon']")
for jobx in job_list:
    title = jobx.find_elements(By.XPATH,".//div/h2//span")[0]
    company=jobx.find_elements(By.XPATH,".//div[@class='heading6 company_location tapItem-gutter companyInfo']//span[@class='companyName']")[0]
    location= jobx.find_elements(By.XPATH,".//div[@class='companyLocation']")[0]
    posting_date = jobx.find_elements(By.XPATH,".//span[@class='date']")[0]
    summary = jobx.find_elements(By.XPATH,".//div[@class='job-snippet']")[0]
    job_info = [title.text, company.text, location.text, summary.text, posting_date.text]
    job_details.append(job_info)
print("====check what inside the job details array=====",job_details)
```

### construct the dataframe after first page web scraping :

```python
dataframe = pd.DataFrame(job_details, columns= ['Job Title', 'Company', 'Location', 'Summary', 'Posting Date'])
dataframe
```

### click on the button for next page:

```python
next_page_btn = driver_chrome_bar.find_element(By.CLASS_NAME, 'pn')
next_page_btn.click()
time.sleep(5)
```

### close the popup:

```python
close_btn= driver_chrome_bar.find_elements(By.XPATH,"//button[@class='icl-CloseButton vjs-x-button-close']")[0]
close_btn.click()
```

### List of all jobs with their elements for the next page and stored object in final job details array :

```python
#list of all jobs with their elements
job_list = driver_chrome_bar.find_elements(By.XPATH,"//div[@class='job_seen_beacon']")
#job_list = driver_chrome_bar.find_elements(By.XPATH,"//div[@class='job_seen_beacon']")
for jobx in job_list:
    title = jobx.find_elements(By.XPATH,".//div/h2//span")[0]
    company=jobx.find_elements(By.XPATH,".//div[@class='heading6 company_location tapItem-gutter companyInfo']//span[@class='companyName']")[0]
    location= jobx.find_elements(By.XPATH,".//div[@class='companyLocation']")[0]
    posting_date = jobx.find_elements(By.XPATH,".//span[@class='date']")[0]
    summary = jobx.find_elements(By.XPATH,".//div[@class='job-snippet']")[0]
    job_info = [title.text, company.text, location.text, summary.text, posting_date.text]
    job_details.append(job_info)
print("===final job details array===",job_details)
```

### construct the final dataframe:

```python
final_df = pd.DataFrame(job_details, columns= ['Job Title', 'Company', 'Location', 'Summary', 'Posting Date'])
```

### close the driver:

```python
driver_chrome_bar.quit()
```
